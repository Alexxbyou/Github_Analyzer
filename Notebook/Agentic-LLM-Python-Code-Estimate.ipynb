```python
# find all python files in folder /Volumes/ORICO/Projects/MLE_Portfolio/04_LLM/04_Agentic_LLM/Agentic-LLM/
# For each file, calculate the total number of lines of code
# Add up all the python files to get the total lines of code in the project
import os
total_lines_of_code = 0
for root, dirs, files in os.walk("/Volumes/ORICO/Projects/MLE_Portfolio/04_LLM/04_Agentic_LLM/Agentic-LLM/"):
    for file in files:
        if file.endswith(".py"):
            with open(os.path.join(root, file), 'r') as f:
                total_lines_of_code += len(f.readlines())
print(f"Total lines of python code in the project: {total_lines_of_code}")
```
